```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### Import the dataset

```python
data = pd.read_csv(r"C:\Users\ksvin\Downloads\Data-Science-Hackathon-main\Data-Science-Hackathon-main\dataset\world-happiness-report-2021.csv")
data.head()
```

### Ladder Score Distribution by Regional Indicator

```python
fig, ax = plt.subplots(figsize=(20, 8))
ax.tick_params(axis='x', rotation=90)

sns.boxplot(x="Regional indicator", y="Ladder score",
            data=data)
```

### Relationship Between Happiness and Income

```python
sns.set_theme(style="ticks")

f, ax = plt.subplots(figsize=(10, 6))
sns.boxplot(x="Logged GDP per capita", y="Regional indicator", data=data,
            whis=[0, 100], width=.6, palette="vlag")

sns.stripplot(x="Logged GDP per capita", y="Regional indicator", data=data,
              size=4, color=".3", linewidth=0)

ax.xaxis.grid(True)
ax.set(ylabel="Regional indicator")
sns.despine(trim=True, left=True)
```
